Calculadora Interativa em Python
Este programa oferece uma calculadora simples e interativa com diversas opções de operações. Ele rodará em um loop contínuo até que você decida sair, exibindo um menu a cada repetição para que você possa escolher a funcionalidade desejada.

```python
# O programa rodará em um loop infinito (while True) até que o usuário decida sair.
# A opção '0' é usada para interromper o loop com um 'break'.
while True:
    # Exibe o menu principal da calculadora a cada iteração do loop.
    print("\n=========================")
    print("**CALCULADORA INTERATIVA**")
    print("=========================")
    print("Digite o número da opção desejada:")
    print("1 - Múltiplos de 5 em um Intervalo")
    print("2 - Contagem de Números Pares em um Intervalo")
    print("3 - Verificador de Número Primo")
    print("4 - Conversor de Temperatura (Cº para Fº)")
    print("0 - Sair do Programa")

    # Solicita a opção do usuário.
    opcao = input("Sua opção: ")

    # ----- Tratamento das Opções (usando if/elif/else) -----
    if opcao == '0':
        # Se o usuário digitar '0', o programa exibe uma mensagem de despedida e encerra.
        print("\nSaindo do programa. Até mais!")
        break # Interrompe o loop 'while True'

    elif opcao == '1':
        # --- Opção 1: Múltiplos de 5 em um Intervalo ---
        print("\n--- Múltiplos de 5 em um Intervalo ---")
        try:
            # Solicita o início e o fim do intervalo e converte para inteiro.
            inicio = int(input("Digite o número inicial do intervalo: "))
            fim = int(input("Digite o número final do intervalo: "))
            print(f"Múltiplos de 5 entre {inicio} e {fim}:")
            encontrou = False # Flag para verificar se algum múltiplo foi encontrado.

            # Percorre o intervalo (incluindo o 'fim').
            for num in range(inicio, fim + 1):
                # Verifica se o número é divisível por 5 (ou seja, múltiplo de 5).
                if num % 5 == 0:
                    print(num)
                    encontrou = True
            if not encontrou:
                print("Nenhum múltiplo de 5 encontrado no intervalo.")
        except ValueError:
            # Captura erros se o usuário digitar algo que não seja um número inteiro.
            print("Entrada inválida. Por favor, digite números inteiros.")

    elif opcao == '2':
        # --- Opção 2: Contagem de Números Pares em um Intervalo ---
        print("\n--- Contagem de Números Pares em um Intervalo ---")
        try:
            # Solicita o início e o fim do intervalo.
            inicio = int(input("Digite o número inicial do intervalo: "))
            fim = int(input("Digite o número final do intervalo: "))
            count_pares = 0 # Inicializa o contador de números pares.
            print(f"Verificando números pares entre {inicio} e {fim}...")

            # Percorre o intervalo e conta os números pares.
            for num in range(inicio, fim + 1):
                if num % 2 == 0:
                    count_pares += 1
            print(f"Total de números pares encontrados: {count_pares}")
        except ValueError:
            print("Entrada inválida. Por favor, digite números inteiros.")

    elif opcao == '3':
        # --- Opção 3: Verificador de Número Primo ---
        print("\n--- Verificador de Número Primo ---")
        try:
            # Solicita um número inteiro positivo.
            numero = int(input("Digite um número inteiro positivo: "))
            if numero <= 1:
                # Números menores ou iguais a 1 não são primos por definição.
                print(f"O número {numero} não é primo.")
            else:
                eh_primo = True # Assume que é primo até que se prove o contrário.
                # O loop vai de 2 até a raiz quadrada do número (otimização).
                for i in range(2, int(numero**0.5) + 1):
                    if numero % i == 0:
                        # Se encontrar um divisor, não é primo.
                        eh_primo = False
                        break # Sai do loop, pois já sabemos que não é primo.
                if eh_primo:
                    print(f"O número {numero} é primo.")
                else:
                    print(f"O número {numero} não é primo.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")

    elif opcao == '4':
        # --- Opção 4: Conversor de Temperatura (Cº para Fº) ---
        print("\n--- Conversor de Temperatura (Cº para Fº) ---")
        try:
            # Solicita a temperatura em Celsius e converte para float.
            celsius = float(input("Digite a temperatura em Celsius (Cº): "))
            # Aplica a fórmula de conversão de Celsius para Fahrenheit.
            fahrenheit = (celsius * 9/5) + 32
            print(f"{celsius:.2f} Cº equivalem a {fahrenheit:.2f} Fº.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número para a temperatura.")

    else:
        # --- Opção Inválida ---
        # Se o usuário digitar algo diferente das opções válidas.
        print("\nOpção inválida. Por favor, escolha uma opção entre 0 e 4.")
```

Como Usar:
Execute a célula de código acima.
O menu da calculadora será exibido.
Digite o número da opção desejada (0, 1, 2, 3 ou 4) e pressione Enter.
Siga as instruções para cada opção.
Para sair, digite 0.
